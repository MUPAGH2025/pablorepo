```python
import pint
import numpy as np
from scipy import constants as const
```

```python
si = pint.UnitRegistry()
```

```python
result_m = (100*si.m+100*si.m+100*si.km) / (100*si.ps**2)
```

```python
result_m
```

```python
result_m.to_base_units()
```

```python
result_m.to(si.km/si.ps**2)
```

How dense is the air?

```python

massMolarDryAir = 28.966 * si.kg / si.kmol
Rd = const.R*si.J/(si.mol*si.K) / massMolarDryAir
print(Rd.to_base_units())


def rho(p, T):
    return p / (Rd * T)
```

```python
rho(1013*si.hPa, 296*si.K).to_base_units()
```

```python
temp_arr = np.arange(18+173, 25+173, 0.2)*si.K
```

```python
rho_arr = rho(1013*si.hPa,temp_arr)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(temp_arr, rho_arr, label='rho w sali')
plt.xlabel('Temperature [K]')
plt.ylabel(f"density")
#plt.gca().xaxis.set_units(si.celsius)
plt.legend()
plt.show()
```
